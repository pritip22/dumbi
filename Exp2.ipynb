```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/students.csv")
print(df)

```

```python
avg = df["Age"].mean()
df["Age"].fillna(avg , inplace = True)

freq = df["Department"].mode()[0]
df["Department"].fillna(freq, inplace = True)
print(df)
```

```python
plt.figure(figsize=(6,6))
plt.bar(df['Department'], df['Marks'], color = 'yellow')
plt.xlabel('Marks')
plt.ylabel('Department')
plt.title('Bar chart of Department vs Marks')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hist(df['Department'], bins = 10, color = 'yellow')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram chart of Department')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Department'], df['Marks'], color = 'yellow')
plt.xlabel('Marks')
plt.ylabel('Department')
plt.title('Scatter chart of Department vs Marks')
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,6))
sns.boxplot(x = 'Department', y = 'Age', data=df, palette = 'Set2')
plt.xlabel('Department')
plt.ylabel('Age')
plt.title('Box chart of Department vs Marks')
plt.tight_layout()

plt.show()
```
